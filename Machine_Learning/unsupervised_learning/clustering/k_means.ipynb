```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Clustering | خوشه بندی

# Generate synthetic data for clustering | تولید داده های مصنوعی برای خوشه بندی
# n_samples=300 → تعداد کل نقاط داده (۳۰۰ نمونه)
# n_features=2 → (y و X محور) هر نمونه دارای ۲ ویژگی 
# centers=3 → تعداد خوشه‌ها (۳ دسته‌ی مختلف از نقاط)
# X → Coordinates of points | شامل مختصات نقاط است
# y → Cluster label for each point | کلاس یا خوشه‌ی هر نقطه را مشخص می‌کند
X, y = make_blobs(n_samples=300, n_features=2, centers=3)

# Plot the data points | رسم نقاط داده روی نمودار
# X[:, 0] → (X محور) مقدار ویژگی اول
# X[:, 1] → (y محور) مقدار ویژگی دوم
plt.scatter(X[:, 0], X[:, 1], color='b')
plt.show()
```

```python
# Create and fit the KMeans model | ساخت مدل کی مینز و آموزش اون

model = KMeans(n_clusters=3) # تعداد خوشه ها
model.fit(X)
```

```python
# Get cluster labels and centers | گرفتن برچسب خوشه‌ ها و مراکز خوشه‌ ها
labels = model.labels_ # نشان می‌دهد هر نقطه‌ی داده ایکس در کدام خوشه قرار گرفته است
centers = model.cluster_centers_ # مراکز خوشه ها

# Plot the clustered data with different colors | رسم نقاط داده با رنگ‌های مختلف بر اساس خوشه
plt.scatter(X[:, 0], X[:, 1], c=labels) # c=labels → هر نقطه را بر اساس خوشه‌ای که در آن قرار گرفته رنگ می‌کند
# Plot the cluster centers with red X | رسم مراکز خوشه‌ها با علامت ایکس قرمز
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.show()
```
