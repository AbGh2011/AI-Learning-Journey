```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Drawing a line plot | کشیدن نمودار خطی

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 3, 7, 8, 1, 3, 9, 11]

plt.plot(x, y, marker='o', color="green") # ایگرگ بر حسب ایکس
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
# Drawing a bar chart | کشیدن نمودار میله ای

x = ["red", "green", "blue", "orange"]
y = [5, 19, 2, 4]

plt.bar(x, y, color="red")
plt.title("Example")
plt.xlabel("color")
plt.ylabel("count")

# Drawing grid lines on the plot | کشیدن خط های شطرنجی روی صفحه
plt.grid()

plt.show()
```

```python
# Drawing a pie chart | کشیدن نمودار دایره ای

x = ["red", "green", "blue", "orange"]
y = [5, 19, 2, 4]

plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red", "green", "blue", "orange"])
plt.title("Example")

plt.show()
```

```python
# Drawing a histogram (showing the number of elements in intervals) | کشیدن نمودار هیستوگرام که تعداد المان ها رو توی یک بازه مشخص نشون میده

x = np.random.rand(1000, 1) # هزار تا عدد از بین 0 تا 1

plt.hist(x, bins=10, color='yellow') # ده تا میله

plt.show()
```

```python
# Drawing a heatmap | کشیدن نمودار هیت مپ

x = np.random.rand(64, 64)

plt.imshow(x, cmap='hot')
plt.colorbar()

plt.show()
```
